# Experiment 1: Scalar Field Axioms

Does the `Scalar` class satisfy all the axioms of a finite field? We test each
axiom with random inputs and track pass/fail results.

**Related:** [Tutorial 1: Scalars and the Finite Field](tutorial-01-scalars.ipynb)

```python
import sys
sys.path.insert(0, "../src")

from frost import Scalar, Q
import secrets

N = 1000  # number of random trials per axiom

def random_scalar():
    return Scalar(secrets.randbits(256) % Q)

def random_nonzero_scalar():
    while True:
        s = random_scalar()
        if int(s) != 0:
            return s

results = {}
print(f"Testing field axioms with N={N} random trials each")
print(f"Field order Q = {Q}")
print(f"Q is prime: {pow(2, Q - 1, Q) == 1} (Fermat test)")
```

## Closure

For any a, b in `Z_Q`: a + b and a·b must also be in `Z_Q` (i.e., valid Scalars
in range [0, Q)).


```python
passed = 0
for _ in range(N):
    a, b = random_scalar(), random_scalar()
    s = a + b
    p = a * b
    if 0 <= int(s) < Q and 0 <= int(p) < Q:
        passed += 1
results["Closure"] = passed
print(f"Closure: {passed}/{N}")
```

## Associativity

(a + b) + c == a + (b + c) and (a·b)·c == a·(b·c).


```python
passed = 0
for _ in range(N):
    a, b, c = random_scalar(), random_scalar(), random_scalar()
    if (a + b) + c == a + (b + c) and (a * b) * c == a * (b * c):
        passed += 1
results["Associativity"] = passed
print(f"Associativity: {passed}/{N}")
```

## Commutativity

a + b == b + a and a·b == b·a.


```python
passed = 0
for _ in range(N):
    a, b = random_scalar(), random_scalar()
    if a + b == b + a and a * b == b * a:
        passed += 1
results["Commutativity"] = passed
print(f"Commutativity: {passed}/{N}")
```

## Identity Elements

a + 0 == a (additive identity) and a·1 == a (multiplicative identity).


```python
zero = Scalar(0)
one = Scalar(1)
passed = 0
for _ in range(N):
    a = random_scalar()
    if a + zero == a and a * one == a:
        passed += 1
results["Identity"] = passed
print(f"Identity: {passed}/{N}")
```

## Inverses

a + (-a) == 0 (additive inverse) and a·a⁻¹ == 1 for nonzero a (multiplicative
inverse).


```python
passed = 0
for _ in range(N):
    a = random_nonzero_scalar()
    if a + (-a) == zero and a * a.inv() == one:
        passed += 1
results["Inverse"] = passed
print(f"Inverse: {passed}/{N}")
```

## Distributivity

a·(b + c) == a·b + a·c.


```python
passed = 0
for _ in range(N):
    a, b, c = random_scalar(), random_scalar(), random_scalar()
    if a * (b + c) == a * b + a * c:
        passed += 1
results["Distributivity"] = passed
print(f"Distributivity: {passed}/{N}")
```

## Results


```python
print(f"{'Axiom':<20} {'Result':<12} {'Status'}")
print("-" * 44)
for axiom, passed in results.items():
    status = "PASS" if passed == N else "FAIL"
    print(f"{axiom:<20} {passed}/{N:<8} {status}")
print()
all_pass = all(v == N for v in results.values())
print(f"All axioms satisfied: {all_pass}")
```

## Why This Works

`Z_Q` is a field because Q is prime. Every nonzero element has a multiplicative
inverse (via Fermat's little theorem: `a^(Q-2)` (mod Q)). If Q were composite,
some nonzero elements would lack inverses (the non-coprime elements), and `Z_Q`
would only be a ring.

This experiment verifies the same properties tested by
`test_properties.py::test_scalar_arithmetic_properties`, but here we see each
axiom individually rather than through Hypothesis's property-based framework.
